## Background and Prior Work

<p>
Sleep quality is a very important part of performing cognitive functions and especially in regards to working memory. Working memory is our brain's ability to store information and use it to reason, make decisions, and learn new things. The prefrontal cortex is a structure in our brain that research in the neuroscience domain has found is activated when working memory is being used. Funahashi (2017)<a href="#fn1" id="ref1-link">1</a> explored the neurological processes that go on in regards to working memory, and he found that the prefrontal cortex acts as the "control center" for working memory, which helps us regulate and manage all the information and stimulus our brain is receiving. When we are sleep deprived, this part of the brain becomes less effective, making it harder to learn and focus.
</p> 

<p> 
Research has proven that poor sleep quality is correlated with lower academic success. For example, Gilbert and Weaver (2010)<a href="#fn5" id="ref5-link">5</a> found a correlation between low GPA scores and poor sleep quality. Poor sleep quality is a very common issue college students face because of scheduling, stress, exams, and workload. 
</p>

<p> 
It has been shown consistently in research that the quality of sleep is correlated with working memory. Frenda and Fenn (2016), in their review of the preexisting literature on sleep deprivation and working memory, found that a lack of sleep reduced the activation in the frontal and parietal lobes, reducing the activation of working memory. Another finding from their work is that a limited number of individuals experience an increase in the activation of neurons when sleep deprived. <a href="#fn2" id="ref2-link">2</a> Their findings suggest that adequate sleep quality is necessary for the proper function of working memory and that these functions are heavily intertwined with the parietal region of the brain. 
</p>
Some researchers have also found that there may not be as much of a correlation between working memory and sleep quality. Almarzouki et al. (2022)<a href="#fn3" id="ref3-link">3</a> did a study on college students and found that during the school year, sleep quality decreased, however, their working memory did not decline based on a cognitive assessment given. This entails that school is correlated with poor sleep quality, but may not always impact working memory. 
</p> 

<p> 
Working memory is also linked to academic performance. E.T. Aronen et al. (2005) studied high school students and found that higher teacher reports were given to students with stronger auditory and spatial working memory. The study also found that students with weaker working memory performance expressed more attention deficits which can lead to weaker learning. <a href="#fn4" id="ref4-link">4</a> The findings can suggest that working memory deficits can contribute to difficulty with learning and academic performance. 
</p> 

<p> 
Taken all together, most of the research supports the idea that working memory function is affected by sleep quality and in turn, affects academic performance. Our project aims to show the relationships between working memory function and sleep quality, and how that will affect college students' academic performance.
</p>

### Prep work for EDA

```python
# Find text columns and standardize them 
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().str.title()

df['sleep_quality'].value_counts()

# Normalize and convert text data into numeric data
sleep_norm = (
    df['sleep_hours']
      .str.replace('–', '-', regex=False)
      .str.replace('—', '-', regex=False)
)

sleep_map = {
    "Less Than 4 Hours": 3,
    "Less Than 4 Hours ": 3,  
    "Less Than 4 Hours": 3,
    "4-5 Hours": 4.5,
    "6-7 Hours": 6.5,
    "7-8 Hours": 7.5,
    "More Than 8 Hours": 9
}

df['sleep_hours_num'] = sleep_norm.map(sleep_map)

gpa_map = {
    "Excellent": 4.0,
    "Good": 3.5,
    "Average": 3.0,
    "Below Average": 2.0,
    "Poor": 1.0
}

df['gpa_num'] = df['academic_performance'].map(gpa_map)

print(df[['sleep_hours', 'sleep_hours_num', 'academic_performance', 'gpa_num']].head())
```

### Scatter Plot of 'Student Insomnia and Educational Outcomes'

```python
# Create Scatter plot and set perimeters 
plt.figure(figsize=(7,5))
plt.scatter(df['sleep_hours_num'], df['gpa_num'])
plt.xlabel('Sleep Hours (numeric)')
plt.ylabel('Academic Performance (GPA-like score)')
plt.title('Sleep Duration vs Academic Performance')
plt.grid(True)
plt.tight_layout()
plt.show()
```

<p>
 The scatterplot shows a weak, but existing positive correlation between sleep duration and academic performance. Students who sleep more tend to have slightly higher GPA-like scores. This trend supports the idea that we hypothesized, that sleep may contribute to cognitive function and academic performance. However, the data also shows some variability which would suggest that sleep alone may not be the only variable predicting academic performance. This could indicate that academic performance, measured by GPA-like scores, is influenced by sleep as well as other factors such as hours spent studying, attendance, etc. Overall, the scatterplot does reveal a slight correlation between sleep and academic performance.
</p>

### Histogram of ''Student Insomnia and Educational Outcomes'

```python
# Create histogram and set perimeters
plt.figure(figsize=(7,5))
plt.hist(df['sleep_hours_num'], bins=5, edgecolor='black')
plt.xlabel('Sleep Hours (numeric)')
plt.ylabel('Number of Students')
plt.title('Histogram of Sleep Duration')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

<p>
The histogram shows that most students get around 7-9 hours of sleep, with relatively less getting under 6 hours. This means most students in the sample get about the amount of sleep that’s usually considered healthy for young adults. Since most students sleep 7–9 hours, very few get very little sleep, which makes it easier to look at how sleep affects grades with fewer potential confounds. This provides a solid baseline to analyze how sleep relates to academic performance. Understanding the typical sleep patterns in this sample allows for a better analysis of how variations in sleep duration, both within the healthy range and at the lower extreme, might influence student outcomes.
</p>

### Boxplot of ''Student Insomnia and Educational Outcomes'

```python
# Create box plot and set perimeters
plt.figure(figsize = (8,6))
df.boxplot(column = 'gpa_num', by = 'sleep_quality')

plt.title('Distribution of Academic Performance by Sleep Quality')
plt.suptitle('')
plt.xlabel('Sleep Quality')
plt.ylabel('Academic Performance (GPA-like Score)')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show
```

<p>
The box and whisker plot illustrates the relationship between students' GPA like academic performance (x-axis) and their reported sleep quality (y-axis), which ranges from "very poor" to "very good". For students who report "average" for their sleep quality, the interquartile range of academic performance is from 1.0-4.0, with its maximum at 4.0. We can very clearly see that the range for "average" sleep quality reaches a higher value than all other sleep quality reportings, which range from 1.0 to 2.0. Additionally, the maximum score for "average" sleep quality reaches 4.0, whereas all other sleep quality reportings' maximums' are at 3.5.
</p>

### Bargraph of ''Student Insomnia and Educational Outcomes'

```python
# Filter and compute average for student with below average GPA 
below_avg = df[df['gpa_num'] == 2.0]

mean_sleep_below_avg = below_avg['sleep_hours_num'].mean()
print('Average sleep hours for Below Average students:', mean_sleep_below_avg)

#Create Bargrpah and set perimeters
plt.figure(figsize = (6,5))
plt.bar(['Below Average'], [mean_sleep_below_avg], color = 'skyblue')
plt.title('Average Sleep Hours for Students With Below Average Academic Performance')
plt.ylabel('Average Sleep Hours')
plt.ylim(0, max(df['sleep_hours_num']) + 1)
plt.tight_layout()
plt.show
```

This suggests that students reporting "average" sleep quality exhibit greater variability in academic performance compared to students in other sleep quality categories. While most students in the other categories cluster within a narrower range of scores (1.0–2.0), the wider range for “average” sleep quality shows that some students are able to achieve higher academic performance despite only average sleep.
